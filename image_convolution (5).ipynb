# Image Convolution

An image in Open CV is simply a matrix. As we learned in our schools it has a width of N and a height of M, so it is known as a 2 dimension n x m matrix. But, in an RGB image we have an additional dimension for the color depth. This additional dimension contains the Red, Green and Blue values for a particular pixel.

The word Convolution is simply a mathematical term, that is basically used with matrices. Since, an image is also a matrix we can use convolution to change some of the characteristics of an image. And that is how Image Convolution comes into the picture.

Image Convolution means changing the matrix of an image with another matrix. The other matrix that we use is called Kernel, Convolution matrix, or mask.

A Kernel is nothing but a small matrix compared to an image matrix. We can have variations of the same image by applying the Image Convolution on an image. For an example, we can blur an image, we can detect the edges in an image and we could even do color detections as well.

```python
#to begin let's import the necessary libraries.
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 200 #resize image on jupyter notebook

#let's create a function for showing 2 images
def showTwoImages(image1, image2, title1, title2):
    fig, (ax1, ax2) = plt.subplots(1,2)
    
    ax1.set_title(title1, size=8)
    ax2.set_title(title2, size=8)
    
    ax1.tick_params(axis='both', labelsize=4) #here i change size of the x and y ticks of ax1
    ax2.tick_params(axis='both', labelsize=4) #here i change size of the x and y ticks of ax2
    
    ax1.imshow(image1)
    ax2.imshow(image2)
    
#this function will help us to show a single image    
def showOneImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()

    
#let's read our image
image_butterfly = cv.cvtColor(cv.imread('photos/butterfly.jpg'), cv.COLOR_BGR2RGB)


showOneImage(image_butterfly, "Original")
```

<b> Before further going, I recommend you to give a glance at Wikipedia page for Image Convolution. Because most of the things that we are going to talk about now are based on Wikipedia.</b>

As we discussed earlier, Image Convolution is a simple mathematical procedure that we can use to find features of an image. The most important thing in Image Convolution is the Kernel. Because the result that you will get after the image convolution can be vary with the shape and the values of the kernel that you used. Choosing the values for the kernel is critical. Fortunately, We have some pre-built functions to do the most common things with images. If you visit the Wikipedia page, you will find few of the kernels that can be used for different outcomes.

```python
#let's start with the basic one

#a kernel is nothing but a small matrix, so let's create a small matrix of 3 x 3
kernel = np.array([
    [0,0,0],
    [0,1,0],
    [0,0,0]
])
#this kernel is called as Identity kernel, because it does nothing and return the same image after the convolution.
#so it is identical to the original

#now, let's apply this kernel to our image

"""
    
    filter2D function takes mainly 3 parameters
    1.) the source image
    2.) ddepth => this is actually the type of the data in the array, -1 means take the same of the original
                  For now, just use -1, because most of the time we do not change the data type. If it is needed
                  to change the data type, then we will dig deeper into this.
    3.) kerel              

"""
identity = cv.filter2D(image_butterfly,-1,kernel)

showTwoImages(image_butterfly, identity, "Original", "Result")
```

<b>Ok great, we applied a kernel but the result was same. And that is because the kernel we used does nothing it is an identity kernel. So, let's try a different kernel this time.

```python
#this kernel is used for edge detection
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])


#now, let's apply this kernel to our image
edged_image = cv.filter2D(image_butterfly,-1,kernel)



showTwoImages(image_butterfly, edged_image, "Original", "Result")
```

<b>Let's try one more kernel

```python
#this kernel is used for bluring -> remember we have to devide it by 9 at the end
kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])/9 #this is similar to np.ones((3,3), np.float32) / 9


#now, let's apply this kernel to our image
blured_image = cv.filter2D(image_butterfly,-1,kernel)

showTwoImages(image_butterfly, blured_image, "Original", "Result")

#as you can see below, the resulting image is blured but it is hard to see, so let's increment our kernel size
```

```python
kernel = np.array(np.ones((20,20), np.float32))/400 


#now, let's apply this kernel to our image
blured_image = cv.filter2D(image_butterfly,-1,kernel)

showTwoImages(image_butterfly, blured_image, "Original", "Result")

#now, it is pretty obvious, how the blured effect is gonna be applied to the result
```
