# Numpy Images

<b> OpenCV reads images as numpy arrays of multi dimensions (ndarrays), by manipulating this array we can change the attributes of an image.

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 200 #resize image on jupyter notebook

#creating my own method for showing the images.
"""

    Using below function, I can show multiple images in the same cell in a row. In order to do 
    that I have used matplotlib pyplot subplots. To look at more details about subplots use the 
    following link to read the official documentation.
    
    https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
    
    And also, I convert the images from BGR to RGB to show it on jupyter, no matter whether 
    the image is already RGB or not.
    
    parameters of the function
    
    1: ) List of images
    2: ) titles for those images
    3: ) list of booleans to indicate whether the image is colored or not
    

"""
def showMultipleImages(images, titles, colored):
    fig, ax = plt.subplots(1,len(images))
    
    for count, cell in enumerate(ax):
        if(colored[count]):
            cell.imshow(images[count])
        else:
            cell.imshow(cv.cvtColor(images[count], cv.COLOR_BGR2RGB)) 
        #need a conversion from bgr to rgb to show it on jupyter    
        
        cell.tick_params(axis='both', labelsize=4) #here i change size of the x and y ticks 
        
        cell.set_title(titles[count]) #giving a title for our images
        
    

#reading the images
colored_image = cv.cvtColor(cv.imread('photos/rgb.png'), cv.COLOR_BGR2RGB) 
#reading the colored image ,Let's convert this from BGR to RGB since it is quite a bit familier

gray_image = cv.imread('photos/rgb.png', cv.IMREAD_GRAYSCALE) 
#reading the grayscale image 

#showing the image
showMultipleImages([colored_image, gray_image], ["color","gray"], [True, False])
    
```

<b> As it is shown, the image contains 3 circles of Red, Green and Blue colors. And we used a matplotlib function to show the image. But, if you directly print the image rather than showing it you will see that the image is actually a numpy array.

```python
#print(image) #This will print a 3 dimensional numpy array, Uncomment the code and run to see

#you can use the np.shape function to look at how many items are there in each dimension
print("COLORED IMAGE SHAPE :- ",colored_image.shape, "\nGRAY IMAGE SHAPE :- ", gray_image.shape)
```

<b> Both the 1st and the 2nd dimensions contain 400 items while the 3rd dimension contains only 3 items. The 1st and 2nd dimensions are the width and height of the image in pixel, so here it is 400 x 400 image. The 3rd dimension represents how many items (colors) in a single pixel, here it is 3 for Red, Green, and Blue. In gray scale images the 3rd dimension is missing, because it does not have RGB values (0 = black, 255 = white).
    <br><br>
<i> To see more details, exmine the following example

```python
print("RED AREA OF THE COLORED IMAGE :- ",colored_image[100,100])
print("GREEN AREA OF THE COLORED IMAGE :- ",colored_image[100,300])
print("BLUE AREA OF THE COLORED IMAGE :- ",colored_image[300,200])
print("BLACK AREA OF THE COLORED IMAGE :- ",colored_image[0,0])
print("BLACK AREA OF THE GRAY IMAGE :- ",gray_image[0,0])
print("GRAY AREA OF THE GRAY IMAGE :- ", gray_image[100,300])

#Uncomment following two lines of codes to see the structure of both the gray and colored images
#print("COLORED IMAGE :- ",colored_image)
#print("GRAY IMAGE :- ", gray_image)
```

<b> Open CV provides us a nice little function to seperate these color channels (Red, Green, Blue) and use them individually. The function is split(), it returns three color channels of an image in a single call.

```python
red, green, blue = cv.split(colored_image)

"""

    Here the split function returns 3 images first image contains only the red color channel,
    so we call it as red, the second image contains only the green color channel so we call it 
    as green, and the third image is for the blue color channel.
    
    what split does is it converts the image into grayscale and makes the corresponding color 
    channel to white and all the other color channels to black. For example in the red 
    color channel image, the area that contains the red color will be white and all the other
    color areas will be black.
    
    So, the sape of these images are 400 x 400 no 3rd dimension.

"""
    
showMultipleImages([red,green, blue],['red','green','blue'], [False, False, False])
```

<b> We can get the actual colors instead of these black and white images, but for that we have to do some extra bit of works. <br><br> Earlier, I said that the colored images have a 3rd dimension, and that 3rd dimesion contains the RGB values. <br><br> Let's say that I want only the red circle of the image, then what I have to is to create a new numpy array that contains 255 values for the red channel while making all the other channels to 0. <br><br> <i> Consider the following code,

```python
zero = np.zeros((400,400),np.uint8)
# creating a new zero channel in addition to the above created red, green, blue channels.
# this zero channel image is totally a black image, since it only contains the value 0 (no 255).

# now let's create new colored images with all Red, Green, Blue color channels using cv.merge function
# remember to asign the zero channel for the color channels that you do not need.

red_circle = cv.merge((red, zero, zero))
green_circle = cv.merge((zero,green,zero))
blue_circle = cv.merge((zero,zero,blue))

showMultipleImages([red_circle,green_circle, blue_circle],['red','green','blue'], [True, True, True])
```
