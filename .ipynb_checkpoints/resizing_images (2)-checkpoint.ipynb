# Resizing images

<b> Image resizing does not work in jupyter notebook. Here all the images are automatically
resized to fit to the screen. So, inorder to resize we use plt.rcParams. So the method 
you are going to see is only applicapable for cv.imshow() method.

```python
import cv2 as cv
from matplotlib import pyplot as plt

plt.rcParams['figure.dpi'] = 200 

image = cv.imread('photos/girl.jpeg') # The image that we are going to resize.

image_1 = cv.cvtColor(image, cv.COLOR_BGR2RGB) 
#Converting color from BGR to RGB format, in order to show it on pyplot.

plt.imshow(image_1)
plt.show()
```

Let's create a simple function that can resize our images.

```python
def resizeImage (img, scale=1):
    width = int(img.shape[1] * scale) #with img.shape we get the actual width and height of the image
    height = int(img.shape[0] * scale) #we can multiply actual width or height by scale
    
    dimensions = (width, height)
    
    return cv.resize(img, dimensions) #return new resized image which is a copy of the original

image_2 = cv.cvtColor(image, cv.COLOR_RGB2BGR)
# Converting color from RGB to BGR again to show it in opencv window.

cv.imshow("original_image", image_2) #Original Image

cv.imshow("resized_image", resizeImage(image_2, 0.5)) #Resized image

cv.waitKey(0)

cv.destroyAllWindows()
```

## Cropping images

<b> In openCv every image is a numpy array. By slicing that numpy array we can crop the image

```python
image_3 = image_1[200:300, 400:800]
# Here the image is cropped with accordance to Y (200 to 300) and X (400 to 800) cordination.

plt.imshow(image_3)
plt.show()
```
