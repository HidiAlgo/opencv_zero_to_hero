## Image Convolution

<b> OpenCV reads images as numpy arrays.

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 200 

image = cv.cvtColor(cv.imread('photos/rgb.png'),cv.COLOR_BGR2RGB) #Let's read our DOG image
#image = cv.imread('photos/rgb.png')

plt.imshow(image)
plt.show()
```

<b> Above image can only be shown by using the imshow method. But if you directly print it to the console, then you will see that the image is actually a numpy array.

```python
red, green, blue = cv.split(image)
zeros = np.zeros(green.shape, np.uint8)
greenBGR = cv.merge((zeros,zeros,blue))

print(image[120][120])
```
