# Let's read an image

<b>To read an image, we use the imread function. with this function we can pass two arguments.
<ol>
    <li> Path to the image </li>
    <li> Flag </li>
</ol>

This flag will specify, how the image should be readed. In the following ways we can read an image.

<ul>
    <li> cv.IMREAD_C0LOR (default mode) -> 1</li>
    <li> cv.IMREAD_GRAYSCALE -> 0</li>
    <li> cv.IMREAD_UNCHANGES -> -1</li>
</ul>
   

```python
import cv2 as cv

image = cv.imread("photos/dog.jpg", cv.IMREAD_COLOR)

cv.imshow("dog", image) #Here the first parameter is the window name.


"""
    This is a keyboard binding function, here this function waits for a given time
    in milisecond. If you pass 0, then it waits for an indefinite period of 
    time until a specific key stroke is pressed.
"""
cv.waitKey(0) 

cv.destroyAllWindows() #This function will destroy all the windows that you created.
```

## Reading and Displaying images with matplotlib

This is the recommended way in jupyter notebook, because the previous method can slow down the kernel. 

```python
import cv2 as cv
from matplotlib import pyplot as plt

"""
    With the following line, the image quality that is rendered on the notebook is imporoved.
    you can adjust the quality by changing the value (200) given.
"""
plt.rcParams['figure.dpi'] = 200 

image = cv.imread("photos/dog.jpg", cv.IMREAD_COLOR)
plt.imshow(image)
plt.show()
```

One other observation that you may observ is that, the colors of the images shown in the plotted section are bit of distorted. This is because opencv reads colors as BGR while matplotlib reads as RGB. So, we should change it.

```python
converted_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(converted_image)
plt.show()
```
